```python
import requests as r
from bs4 import BeautifulSoup
import pandas as pd
import sqlalchemy

url = 'https://buy.yungching.com.tw/region/%E5%8F%B0%E5%8C%97%E5%B8%82-_c/?pg='
url_list = []
for x in range(1,290):
    url_list.append(url+str(x))

final_content = []
for y in url_list:
    res=r.get(y,headers = headers)
    soup = BeautifulSoup(res.text,'lxml')
    crawling_list = soup.find_all('li',class_='m-list-item')

    for parse_article in crawling_list:
        p_dict={}
        p_dict['address'] = parse_article.select('.item-info')[0].h3.text.split()[1]
        p_dict['place'] = parse_article.select('.item-info-detail')[0].text.split('\n')[7]
        p_dict['price'] = parse_article.select('.price')[0].text
        final_content.append(p_dict)
        print p_dict['price']
        
    final = pd.DataFrame(final_content)    #pd,DataFrame(裡面為list)
    engine = sqlalchemy.create_engine('mysql+mysqldb://root:password@localhost/Yongqing?charset=utf8')
    final.to_sql('house', engine, if_exists='replace', index=False, index_label=final)
    
```
