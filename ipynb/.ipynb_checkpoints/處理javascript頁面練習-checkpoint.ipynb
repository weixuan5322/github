```python
from selenium import webdriver#driver的API
from selenium.webdriver.common.keys import Keys   #輸入值的API
```

```python
driver = webdriver.Firefox()
driver.get("http://www.python.org")
elem=driver.find_element_by_id('id-search-field')#以id值找尋input框
elem.clear()#清除
elem.send_keys('selenium')#再輸入要找的值
elem.send_keys(Keys.RETURN)#enter
html_source=driver.page_source#Gets the source of the current page. 取得搜尋結果
print html_source
driver.close()
```

# chromedriver.exe

```python
from selenium import webdriver
from selenium.webdriver.support.ui import WebDriverWait #等待結束的API
from selenium.webdriver.support import expected_conditions #設定條件api
from selenium.webdriver.common.by import By
```

```python
driver = webdriver.Chrome("C://chromedriver.exe")#從哪裡讀取chromedriver.exe
driver.get("https://www.agoda.com/zh-tw/")
driver.find_element_by_xpath("//input[@type='text']").click()#直接到右鍵檢查你所要的位置(post position)，再到下面html案copy得到xpath
driver.find_element_by_xpath("//input[@type='text']").clear()
driver.find_element_by_xpath("//input[@type='text']").send_keys(u"舊金山")
driver.find_element_by_xpath('//*[@id="oneline-searchbox"]/dl[5]/dd/button').click()#點選搜出好價格

elem_present = expected_conditions.presence_of_element_located((By.CLASS_NAME, 'hotel-name'))
#等待條件為By.CLASS_NAME('hotel-name')出現
WebDriverWait(driver, 10).until(elem_present)
#等待10秒後直到'hotel-name'出現

for x in range(0,9):#會寫0到8跑9次是因為已經知道有9頁 這裡也可以用while迴圈跑不知道幾頁
    elem = driver.find_element_by_id("paginationNext")#下一頁的button
    driver.execute_script("return arguments[0].scrollIntoView();", elem)#滾輪滾到最底
    elem.click()
    

print driver.page_source
```

# Regular Expression (正規表達式)
- a-z A-Z 0-9: 一對一match
- . : 任意字元除\n
- \w : a-z A-Z 0-9
- \d : 0-9
- ^ : start
- $ : end
- In [23]:


```python
import re
```

```python
mystr = "I love pikachu, he is really cute"
```

```python
match = re.search(r'pikachu', mystr)
```

```python
print match.group()
```

```python
# + -> repeat previous char 1 to n times
# * -> repeat previous char 0 to n times
# ? -> 0 or 1 time
```

```python
mypet = 'dddddd gdddd'
myurl = "https://google.com/"
match = re.search(r'\w', mypet)
match = re.search(r'\w+', mypet)
match = re.search(r'\w+ g?', mypet)
print match.group()
```

```python
myemail = 'asdfjasdi asas  39402jsaf ianchen06@gmail.com nice nice pika pika'
```

```python
match = re.search(r'\w+@[\w.]+', myemail)
```

```python
match.group()
```

```python
a = '他欠我 250,000元, 好熱阿'
b = 'a欠我 4,000元, 好熱阿'
c = 'B欠我 250元, 好熱阿'

# {n} repeat n times
m = re.search('\d{3}', c)
m = re.search('\d\d\d', c)

print m.group()
```

```python
m = re.search('\d{1,3}', c)
m = re.search('\d{1,3}', b)

print m.group()
```

```python
m = re.search('\d{1,}', c)
print m.group()
```

```python
m = re.search('\d+', c)
print m.group()
```
