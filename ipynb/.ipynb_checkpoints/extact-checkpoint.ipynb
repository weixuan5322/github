## BeautifulSoup


```python
res3 = requests.get("http://www.appledaily.com.tw/column/index")
```

```python
print res3.text
```

```python
from bs4 import BeautifulSoup
```

```python
soup = BeautifulSoup(res3.text,'lxml')#把get下來的網頁進行 extract teg
```

```python
ul_rtddd = soup.find_all('ul',class_='list blue')
ul_rtddd#回傳值動作
```

```python
type(ul_rtddd)#已經得到ul teg物件=ResultSet是物件
```

```python
ul_rtddd_tag =ul_rtddd[0]#找尋其ul的內容值 才能的到tag
type(ul_rtddd_tag)
```

```python
dir(ul_rtddd_tag) #dir後在知道用哪些方法可以取值
```

```python
article_links_list = ul_rtddd_tag.find_all('a')
```

```python
len(article_links_list)
```

```python
result_list = []
for link in article_links_list:#把link值作迴圈到 有30筆就列印30次
    print link['href']
    result_list.append(link['href'])#做了一個list讓其他function可以使用
    
```
