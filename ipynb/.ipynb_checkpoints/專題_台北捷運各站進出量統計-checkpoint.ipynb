```python
import requests as r
from bs4 import BeautifulSoup

url = 'http://data.taipei/opendata/datalist/datasetMeta?oid=1d71c478-205f-42c5-8386-35f86d74fdd1'
headers = {'User-Agent':'Mozilla/5.0 (X11; Ubuntu; Linux x86_64; rv:49.0) Gecko/20100101 Firefox/49.0'}
cookies = {'JSESSIONID':'D4C23787AFAEFF5778882F2BC69CCD71'}
res = r.get(url,headers = headers,cookies = cookies)
soup = BeautifulSoup(res.text,'lxml')
```

```python
soup
```

```python
print soup.find_all('a',class_='resource-url-analytics')[1]['href']
print soup.find_all('a',class_='heading')[0]['title']
```

```python
count = 0

for x in range(0,218):
    if x % 2 == 1:
        print 'http://data.taipei/'+soup.find_all('a',class_='resource-url-analytics')[x]['href']
        print soup.find_all('a',class_='heading')[0]['title']
    count = count +1
    print count
```

```python
url_list = []
for x in range(0,count):
    url_dict = {}
    if x % 2 == 1 :       
        url_dict["url"]='http://data.taipei'+soup.find_all('a',class_='resource-url-analytics')[x]['href']
        url_dict["title"]=soup.find_all('a',class_='heading')[0]['title']
        url_list.append(url_dict)
    
```

```python
print url_list[0]['url']
```

```python
for save in url_list:
    res_2=r.get(save['url'],headers =headers)
    load = '/home/user/Desktop/北市公開資料/'+save['title'].encode('utf8')+'.csv'
    with open(load, 'w') as f:
        f.write(res_2.content)
```
