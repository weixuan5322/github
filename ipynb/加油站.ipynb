```python
import requests as r
from bs4 import BeautifulSoup
import pandas as pd
import sqlalchemy


url = 'http://data.taipei/opendata/datalist/datasetMeta/download?id=2385447a-6793-4055-8ce0-b601ffa5e9b5&rid=a8fd6811-ab29-40dd-9b92-383e8ebd2a4e'
final = []
headers = {'User-Agent':'Mozilla/5.0 (X11; Ubuntu; Linux x86_64; rv:49.0) Gecko/20100101 Firefox/49.0'}
res = r.get(url,headers = headers)
soup=BeautifulSoup(res.text,'lxml')


for x in range(0,78):
    art_dict={}
    art_dict['addr']=soup.select('address')[x].text
    art_dict['x'] = soup.select('addr_x')[x].text
    art_dict['y'] = soup.select('addr_y')[x].text
    art_dict['zone']=soup.select('cityzone')[x].text
    art_dict['name']=soup.select('name')[x].text
    final.append(art_dict)
    
final_pd = pd.DataFrame(final)
engine = sqlalchemy.create_engine('mysql+mysqldb://root:root@localhost/traffic?charset=utf8')
final_pd.to_sql('oil_gas', engine, if_exists='replace', index=False, index_label=final)
```
