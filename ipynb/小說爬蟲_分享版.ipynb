## 內文爬蟲
- 以dictionary作架構
- 把要的部份,find_all取出標籤所在的位置，次之確認標籤後進行取值
- 確認該{ [ ] : [ ] } list所對應的值 以dirctionary 架構下 key:value

```python
def parse_article(url):
    art_dict={}
    res=r.get(url)
    res.encoding='big5'
    soup=BeautifulSoup(res.text,'lxml')
    art_dict['title']=soup.find('h1',class_='novel_title').text
    art_dict['content']=soup.find_all('div',class_='novel_content')[0].text
    art_dict['category']=soup.find_all('p',class_='novel_info')[0].text
    return art_dict
```

# 自動抓取內文及換頁

```python
url="http://tw.piaotian.cc/read/25640/" #url前半段
content_article=[]
for page in range (0,5):#爬取4篇
    
    sum_page=8117292+page    
    crawling_title= url+str(sum_page)+".html"
    #print parse_article(crawling_title)['title']    
    #print parse_article(crawling_title)['content']
    content_article.append(parse_article(crawling_title))
    
print content_article #確認crowling title數量對與否
```

## 存進資料庫
- step 1 :建立資料庫時必須以utf8做為儲存類型 避免文字出現亂碼
- step 2 :把爬下來的資料(content_airticle) type為list 裡面為dictionary 進行取得 key:value 動作
- step 3 :format時需把文字型態轉為utf-8

```python
db =MySQLdb.connect('localhost','root','Password','databaseName',charset='utf8')
cursor=db.cursor()#執行sql語法套件
for count in range(0,len(content_article)):#用迴圈把我們剛剛的每一篇存進table裡面
    title2=content_article[count].values()[2]  #取標題字串
    content2=content_article[count].values()[0] #取內容字串
    category2=content_article[count].values()[1] #取作者字串
    sql = """INSERT INTO story2 (title, content, category) 
            VALUES ("{}", "{}", "{}")""".format(title2.encode("utf-8"),content2.encode("utf-8"),category2.encode("utf-8"))
    cursor.execute(sql)
    db.commit()#確認sql指令 資料就會存進database裡
db.close()
```
